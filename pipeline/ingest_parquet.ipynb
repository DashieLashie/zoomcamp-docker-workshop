```python
import pandas as pd
```

```python
url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-11.parquet'
df = pd.read_parquet(url)
df
```

```python
df[df['trip_distance'] <=1]
```

```python
# rows where pickup is on 2025-11-18
df_nov18 = df[
    (df['tpep_pickup_datetime'] >= '2025-11-18') &
    (df['tpep_pickup_datetime'] <  '2025-11-19')
]
trips_per_zone = (
    df_nov18
    .groupby('PULocationID')
    .size()
    .reset_index(name='trip_count')
    .sort_values('trip_count', ascending=False)
)
trips_per_zone
```

```python
east_harlem_north_df = df[
    (df['PULocationID'] == 74)
]
east_harlem_north_df.sort_values('tip_amount')
```
